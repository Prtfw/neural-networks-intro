## Learning A Shallow Neural Network using Gradient Descent with `numpy`

To learn the weights and biases of our shallow neural-network, we'll first need some data. There are two functions in `datasets.general` that we'll use to generate some fake data, and then we'll demonstrate that our neural network is capable of learning the functions that generated this data. 

For the most part, the code below is going to resemble the equations that we've looked at so far. Probably the biggest difference is that we'll create weight and bias vectors per layer, rather than creating individual variables for each weight and bias (see the [appendix](https://github.com/sallamander/neural-networks-intro/blob/master/mini-books/shallow-neural-networks/04-shallow-neural-network/appendix.ipynb) for a non-vector version of the code). This will make the code a little cleaner and a little faster. It's also how standard neural network libraries operate, and it'll be good to start getting used to it while the networks we're building are relatively simple.  

```python
import numpy as np
import matplotlib.pyplot as plt
from datasets.general import gen_trigonometric, gen_powx
from utils.plotting import plot_func
from activations import sigmoid
%matplotlib inline
```

```python
def learn_w_gradient_descent(xs, ys, hlayer_size=2):
    learning_rate = 0.001
    
    #####################################################################
    ##### Step 1 - Initialize weights and biases with random values #####  
    #####################################################################
    
    ##### Weights/Biases connecting Layer 1 to Layer 2 #####
    W2 = 2 * np.random.random(size=(1, hlayer_size)) - 1 
    b2 = np.zeros(hlayer_size)   
    
    ##### Weights/Biases connecting Layer 2 to Layer 3 #####
    W3 = 2 * np.random.random(size=(hlayer_size, 1)) - 1
    b3_1 = 0

    for _ in range(50000):
        
        #####################################################
        ##### Step 2A - calculate our predicted values. #####
        #####################################################
        
        ##### Layer 2 calculations ##### 
        z2 = xs.dot(W2) + b2 # weighted sums into Layer 2         
        a2 = sigmoid(z2) # activations from Layer 2 
        
        ##### Layer 3 (output layer) calculations ##### 
        yhats = a2.dot(W3) + b3_1 # weighted sum into Layer 3
        
        ###########################################
        ##### Step 2B - calculate our errors. #####
        ###########################################
        diffs = ys - yhats 
        es = 0.5 * (diffs ** 2) # not currently used
        
        ######################################################################
        ##### Step 2C - calculate the gradient of the error with respect #####
        ##### to the weights/biases, and use it to update the coefficients. ##
        ######################################################################
        
        ##### Gradient/derivative calculations #####
        db3_1 = -diffs
        dW3 = -diffs * a2 
        dW2 = np.dot((-diffs * W3.T * a2 * (1 - a2)).T, xs)
        db2 = -diffs * W3.T * a2 * (1 - a2)
        
        ##### Updates ##### 
        W3 -= learning_rate * dW3.mean(axis=0).reshape(-1, 1) 
        b3_1 -= learning_rate * db3_1.mean(axis=0)
        W2 -= learning_rate * dW2.mean(axis=0).reshape(-1, 1)
        b2 -= learning_rate * db2.mean(axis=0)
    
    # This will hold our predictions from the final iteration
    return yhats
```

You'll notice above that we've included a parameter in our function for the hidden layer size (number of nodes). In our computational graphs and derivations, we've been working with two, but it's not too hard to scale up. In order to do so, we simply have to adjust the size of our weight vectors for layers 2 and 3 (`W2` and `W3`), and the bias vector for layer 2 (`b2`). This is done using the `hlayer_size` parameter. The derivative calculations for additional nodes involve the same chain rule we used before, so as long as we're careful to line up the right pieces, the derivative calculations will also scale.

Why more nodes, though? Well, it turns out that a neural network with one hidden layer having two nodes isn't too powerful. Making the network **wider** (e.g. adding more nodes) gives the network more representational power, and allows it to more effectively model the relationships underlying the data that it's looking at. Let's take a look at this in action by trying to have our network learn two functions: <img src="../imgs/equations/y_sinx.png" style="vertical-align: text-middle; display: inline-block; padding-top:0; margin-top:0;" width=85\>
and 
<img src="../imgs/equations/y_x4.png" style="vertical-align: text-middle; display: inline-block; padding-top:0; margin-top:0; padding-bottom:0;" width=55\>. 

```python
np.random.seed(317) # set seed for reproducibility 
# Randomly generate some fake data that follows a y = sin(x) relationship.
n_obs = 50
xs, ys = gen_trigonometric(n_obs=n_obs, func='sine')

# Fit four different sized networks, plotting the true function and 
# network generated function (e.g. it's predictions) for each network
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
hlayer_sizes = (2, 10, 100, 250)
for ax, hlayer_size in zip(axes.flatten(), hlayer_sizes):
    yhats = learn_w_gradient_descent(xs, ys, hlayer_size=hlayer_size)
    plot_func(xs=xs, y_true=ys, y_pred=yhats, n_nodes=hlayer_size, func='sine', ax=ax)
```

```python
np.random.seed(317) # set seed for reproducibility 
# Randomly generate some fake data that follows a y = x^4 relationship.
n_obs = 50
xs, ys = gen_powx(n_obs=n_obs, pow_x=4)

# Fit four different sized networks, plotting the true function and 
# network generated function (e.g. it's predictions) for each network
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
hlayer_sizes = (2, 10, 100, 250)
for ax, hlayer_size in zip(axes.flatten(), hlayer_sizes):
    yhats = learn_w_gradient_descent(xs, ys, hlayer_size=hlayer_size)
    plot_func(xs=xs, y_true=ys, y_pred=yhats, n_nodes=hlayer_size, func='X^4', ax=ax)
```

With both functions, we see the same pattern - as we add more nodes to the hidden layer, the network is able to more accurately represent the function. In fact, if we were to keep adding more and more nodes, the network would eventually be able to model these functions with incredibly high levels of accuracy (if it's not already). This extends to other functions as well, and this allows us to know that neural networks have **universality**. In other words, they can basically learn to model any function within a reasonable level of accuracy. We can test this out by playing around with other values for the `func` argument in `gen_trigonometric` (acceptable values are *sine*, *cosine*, or *tanh*) or the `pow_x` argument in `gen_powx` (any integer value there is fair game, although high ones will throw overflow errors). 

This **universality** is what makes neural networks so powerful! Later on, we'll see this power and flexiblity extend to real world data sets (rather than toy mathematical functions), where we'll be able to achieve astounding results on regression and classification problems.
